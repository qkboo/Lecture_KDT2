# 06. Groupby와 Pivot table(with타이타닉호 데이터)

```python
df1 = df.copy()
```

### apply() - 함

```python
df1['who'].value_counts()
```

```python
def transform_who(x):
    if x == 'man':
        return '남자'
    elif x == 'woman':
        return '여자'
    else:
        return '아이'
```

```python
df1['who'].apply(transform_who )
```

```python
df1['who'] = df1['who'].apply(transform_who )
df1['who'].value_counts()
```

### lambda

```python
df1['survived'].value_counts()
```

```python
df1['survived'].apply(lambda x: '생존' if x == 1 else '사망')
```

```python
df1['survived'] = df1['survived'].apply(lambda x: '생존' if x == 1 else '사망')
df1['survived'].value_counts()
```

## group by

```python
df1.groupby(['sex','pclass'])  # 성별로 객실 클래스를 묶는다.
```

groupby()를 사용할 때는 반드시 aggregate 하는 통계함수와 일반적으로 같이 적용합니다.

```python
df1.groupby(['sex','pclass']).mean()  # 성별로 객실 클래스를 묶는다.
```

```python
df.groupby(['sex','pclass']).agg('sum', 'count')
```

```python
df.groupby(['sex','pclass'])['survived'].agg('sum', 'count')
```

```python
df.groupby(['sex','pclass'])['survived'].mean()
```

### reset_index()

```python
df1.head(3)
```

```python
df1.reset_index()
```

```python
df1.reset_index(drop=True)
```

```python
df.groupby(['sex','pclass']).agg('sum', 'count')
```

```python
df.groupby(['sex','pclass']).agg('sum', 'count').reset_index()
```

### pivot

```python
# index에 묶을 그룹 / values = 선택 컬럼
df.pivot_table(index='who', values='survived')
```

```python
df.pivot_table(index=['sex','pclass'], values='survived')
```

```python
df.pivot_table(index=['sex','pclass'], values='survived',
              aggfunc=['sum', 'mean'])
```

```python
df.pivot_table(index=['sex','pclass'], values='survived',
              aggfunc=['sum', 'mean']).reset_index()
```
